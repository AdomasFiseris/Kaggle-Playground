### Playground Series S5E5 EDA

```python
import pandas as pd
import numpy as np
import seaborn as sns
import kaggle
```

```python
#!kaggle competitions files -c playground-series-s5e5
#!kaggle competitions download -c playground-series-s5e5
```

```python
train = pd.read_csv('playground-series-s5e5/train.csv')
train.info()
```

```python
num_columns = [col for col in train.columns if col not in ['id', 'Sex']]
train[num_columns].hist(figsize=(12,8), bins=40)
```

```python
sns.countplot(x='Sex', data=train)
```

```python
sns.heatmap(train[num_columns].corr(method='spearman'), annot=False, cmap='coolwarm', square=True)
```

```python
sns.pairplot(train[:10000], vars=num_columns, hue='Sex', corner=True)
```

### Feature Engineering

```python
train['BMI'] = (train['Weight'] / (train['Height'] / 100) ** 2).round(2)
train['BMI'].value_counts().head(15).reset_index()
```
